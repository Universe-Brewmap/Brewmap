```python
# 1. 테스트용 데이터 준비 (100개만 테스트)
test_df = master_df.head(100).copy() 
school = df_list[14]

# 2. 시설 개수를 세는 함수
def count_nearby_facilities(row, facilities_df, radius=0.003):
    # 카페의 좌표
    lat = row['lat']
    lng = row['lng']
    
    # 시설물의 좌표가 반경 내에 있는 것들만 필터링
    nearby_facilities = facilities_df[
        ((facilities_df['lat'] - lat)**2 + (facilities_df['lng'] - lng)**2) <= radius**2
    ]
    
    return len(nearby_facilities)

# 3. 적용 (각 카페 행마다 함수 실행)
print("시설 수 계산 중...")
test_df['academy_count'] = test_df.apply(lambda row: count_nearby_facilities(row, school), axis=1)

print("결과 확인:")
display(test_df[['nm', 'addr', 'academy_count']].head())
```

```python
from tqdm import tqdm
import pandas as pd

# 1. 계산할 시설물들을 딕셔너리에 담습니다. (key는 결과 컬럼명, value는 시설물 데이터프레임)
facilities_map = {
    'school_count': school,
    'bus_stop_count': bus_stop,
    'crosswalk_count': crosswalk,
    'gym_count': gym_df,
    'hospital_count': hospital_df,
    'physical_count': physical_df,
    'academy_count': academy_df,
    'industry_center_count': industry_center
}

# 2. 한 번에 모든 시설물을 계산하는 함수 정의
def count_all_facilities(row):
    lat, lng = row['lat'], row['lng']
    radius = 0.003 # 300m 반경 설정
    
    results = {}
    for col_name, facility_df in facilities_map.items():
        # 거리 계산 로직 (기존 함수 재사용)
        # 시설물의 lat, lng와 매물의 lat, lng를 비교
        count = len(facility_df[
            ((facility_df['lat'] - lat)**2 + (facility_df['lng'] - lng)**2) <= radius**2
        ])
        results[col_name] = count
    
    # 딕셔너리를 Series로 변환하여 반환 (Pandas가 자동으로 컬럼에 매핑해줍니다)
    return pd.Series(results)

# 3. 진행 상황을 보여주며 전체 데이터에 한 번에 적용
tqdm.pandas()
print("모든 시설물을 한 번에 계산 중입니다...")

# 새로운 컬럼들을 한꺼번에 생성
master_df[list(facilities_map.keys())] = master_df.progress_apply(count_all_facilities, axis=1)

# 4. 완료 후 반드시 저장!
master_df.to_csv('master_df_final_features.csv', index=False, encoding='utf-8-sig')
print("병합 및 카운팅 완료! 파일 저장")
```

# 바나프레소 데이터 추출 라벨링

```python
import pandas as pd
from datetime import datetime

# 1. 아까 저장한 마스터 데이터 불러오기 (만약 이미 켜져 있다면 생략 가능)
# master_df = pd.read_csv('master_df_final_features.csv', encoding='utf-8-sig')

# 2. '바나프레소' 매장만 필터링하기
# 카페 이름(nm) 컬럼에 '바나프레소'가 포함된 데이터만 찾습니다.
bana_df = master_df[master_df['nm'].str.contains('바나프레소', na=False)].copy()
print(f"찾아낸 바나프레소 매장 수: {len(bana_df)}개")

# 3. 영업 기간 계산 및 라벨링 (Target 만들기)
# apv_date(인허가일자)를 기준으로 현재까지 몇 년이 지났는지 계산합니다.
# (만약 데이터에 폐업 여부 컬럼이 따로 있다면 그 기준을 합쳐야 합니다)

# 날짜 형식으로 변환 (YYYY-MM-DD)
bana_df['apv_date'] = pd.to_datetime(bana_df['apv_date'], errors='coerce')

# 오늘 날짜 기준으로 영업 일수 계산 후 '년(year)' 단위로 변환
today = pd.to_datetime('today')
bana_df['operating_years'] = (today - bana_df['apv_date']).dt.days / 365

# 3년 이상이면 1(성공), 아니면 0(실패)
bana_df['target'] = bana_df['operating_years'].apply(lambda x: 1 if x >= 3 else 0)

# 결과 확인
print("\n라벨링 완료! 성공(1)과 실패(0) 매장 수:")
print(bana_df['target'].value_counts())
display(bana_df[['nm', 'apv_date', 'operating_years', 'target']].head())

# 1. 이름(nm)과 주소(addr)가 완전히 똑같은 중복 데이터들을 하나만 남기고 다 지웁니다.
bana_real_df = bana_df.drop_duplicates(subset=['nm', 'addr']).copy()

# 2. 뻥튀기가 빠진 '진짜' 바나프레소 매장 수 확인
print(f"진짜 바나프레소 매장 수: {len(bana_real_df)}개")

# 3. 진짜 라벨링 결과 확인
print("\n중복 제거 후 성공(1)과 실패(0) 매장 수:")
print(bana_real_df['target'].value_counts())
display(bana_real_df[['nm', 'apv_date', 'operating_years', 'target']].head())
```

```python
import numpy as np
from math import radians

# 직방 매물 다시 불러오기
property_df = pd.read_csv("E:/web_prj/Brewmap_api/data/mj/test_zigbang.csv", encoding="utf-8-sig")

# 조건 1. 1층, 지하1층, 반지하층만 필터링
property_df = property_df[
    property_df['floor'].isin(['1층', '지하1층', '반지하층'])
]
print("층수 필터 후 매물 수:", len(property_df))

# 조건 2. 반경 800m 내 지하철역 또는 버스정류장
def has_nearby_transit_fast(lat, lng, subway_df, bus_df, radius=800):
    lat1, lng1 = radians(lat), radians(lng)
    R = 6371000

    # 지하철 확인
    sub_lats = np.radians(subway_df['lat'].dropna().values)
    sub_lngs = np.radians(subway_df['lng'].dropna().values)
    dlat = sub_lats - lat1
    dlng = sub_lngs - lng1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(sub_lats) * np.sin(dlng/2)**2
    if (R * 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a)) <= radius).any():
        return True

    # 버스 확인
    bus_lats = np.radians(bus_df['lat'].dropna().values)
    bus_lngs = np.radians(bus_df['lng'].dropna().values)
    dlat = bus_lats - lat1
    dlng = bus_lngs - lng1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(bus_lats) * np.sin(dlng/2)**2
    if (R * 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a)) <= radius).any():
        return True

    return False

# 필터 적용
property_df['pass_filter'] = property_df.apply(
    lambda x: has_nearby_transit_fast(x['lat'], x['lng'], df_dh, bus_stop),
    axis=1
)

# 필터 통과한 매물만
property_df = property_df[property_df['pass_filter'] == True]
print("최종 필터 후 매물 수:", len(property_df))
```

```python
print("시설 수 카운팅 시작...")

property_df["학교수"] = property_df.apply(lambda x: count_nearby_fast(x['lat'], x['lng'], school), axis=1)
property_df["병원수"] = property_df.apply(lambda x: count_nearby_fast(x['lat'], x['lng'], hospital_df), axis=1)
property_df["학원수"] = property_df.apply(lambda x: count_nearby_fast(x['lat'], x['lng'], academy_df), axis=1)
property_df["체육도장수"] = property_df.apply(lambda x: count_nearby_fast(x['lat'], x['lng'], physical_df), axis=1)
property_df["지식산업센터수"] = property_df.apply(lambda x: count_nearby_fast(x['lat'], x['lng'], industry_center), axis=1)

# 접근성 - 최소 거리 계산
def get_min_distance(lat, lng, facility_df):
    lats = np.radians(facility_df['lat'].dropna().values)
    lngs = np.radians(facility_df['lng'].dropna().values)
    lat1, lng1 = radians(lat), radians(lng)
    dlat = lats - lat1
    dlng = lngs - lng1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lats)*np.sin(dlng/2)**2
    distances = 6371000 * 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return distances.min()

property_df["지하철최소거리"] = property_df.apply(lambda x: get_min_distance(x['lat'], x['lng'], df_dh), axis=1)
property_df["버스최소거리"]   = property_df.apply(lambda x: get_min_distance(x['lat'], x['lng'], bus_stop), axis=1)
property_df["횡단보도수"]     = property_df.apply(lambda x: count_nearby_fast(x['lat'], x['lng'], crosswalk), axis=1)

print("카운팅 완료!")
print(property_df.shape)
```

## MIN-MAX

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# 1. 수요 점수 (25점)
demand_cols = ["학교수", "병원수", "학원수", "체육도장수", "지식산업센터수"]
property_df[demand_cols] = scaler.fit_transform(property_df[demand_cols])
property_df['수요점수'] = property_df[demand_cols].mean(axis=1) * 25

# 2. 접근성 점수 (25점)
# 거리는 작을수록 좋으니까 반전
property_df[['지하철최소거리', '버스최소거리']] = scaler.fit_transform(
    property_df[['지하철최소거리', '버스최소거리']]
)
property_df['지하철최소거리'] = 1 - property_df['지하철최소거리']
property_df['버스최소거리']   = 1 - property_df['버스최소거리']

property_df[['횡단보도수']] = scaler.fit_transform(property_df[['횡단보도수']])
access_cols = ['지하철최소거리', '버스최소거리', '횡단보도수']
property_df['접근성점수'] = property_df[access_cols].mean(axis=1) * 25

# 3. 비용 리스크 점수 (25점)
# 평당 월세 계산
property_df['평당월세'] = property_df['monthly_rent'] / (property_df['size_m2'] / 3.3)

# 행정동별 평균 월세
property_df['월세지수'] = property_df['monthly_rent'] / property_df.groupby('address')['monthly_rent'].transform('mean')

cost_cols = ['평당월세', '월세지수']
property_df[cost_cols] = scaler.fit_transform(property_df[cost_cols])

# 비용은 낮을수록 좋으니까 반전
property_df['비용리스크점수'] = (1 - property_df[cost_cols].mean(axis=1)) * 25

# 4. 최종 점수
property_df['최종점수'] = (
    property_df['수요점수'] +
    property_df['접근성점수'] +
    property_df['비용리스크점수']
)

# Top-3 추천
top3 = property_df.nlargest(3, '최종점수')
print(top3[['address', 'floor', '수요점수', '접근성점수', '비용리스크점수', '최종점수']])
```
