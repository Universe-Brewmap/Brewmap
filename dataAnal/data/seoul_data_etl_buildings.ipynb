```python
import time
import json

apikey = os.getenv("MY_API_KEY")
all_data = [] # 모든 데이터를 모아둘 리스트

# 1. 전체 데이터 개수 확인
first_url = f"http://openapi.seoul.go.kr:8088/{apikey}/json/neisSchoolInfo/1/1/"
response = requests.get(first_url).json()
total_count = response['neisSchoolInfo']['list_total_count']
print(f"총 데이터 개수: {total_count}")

# 2. 500개씩 끊어서 반복문(for)을 돌립니다.
step = 500
for i in range(1, total_count + 1, step):
    start_index = i
    end_index = i + step - 1
    
    # 마지막 데이터 범위를 넘지 않게 조절
    if end_index > total_count:
        end_index = total_count
        
    url = f"http://openapi.seoul.go.kr:8088/{apikey}/json/neisSchoolInfo/{start_index}/{end_index}/"
    
    try:
        response = requests.get(url).json()
        # 데이터가 있는 경우 리스트에 추가
        if 'neisSchoolInfo' in response and 'row' in response['neisSchoolInfo']:
            all_data.extend(response['neisSchoolInfo']['row'])
            print(f"성공: {start_index} ~ {end_index} 가져오기 완료!")
        
        # 서버에 무리가 가지 않게 아주 잠깐 멈춰주는 센스!
        time.sleep(0.1) 
        
    except Exception as e:
        print(f"에러 발생: {e}")

print(f"최종적으로 가져온 데이터 개수: {len(all_data)}")
```
